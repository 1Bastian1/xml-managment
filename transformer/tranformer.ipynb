```python
import pandas as pd
```

```python
df_main = pd.read_csv("./data_demo/MAIN.csv")
df_main.head()
```

CREACION DE DICCIONARIO PARA PROVEEDORES

```python
dict_proveedores = df_main['proveedor'].to_dict()

dict_proveedores_sin_repetido = {}

for i, (key, value) in enumerate(dict_proveedores.items()):
    if value not in dict_proveedores_sin_repetido.values() :
        dict_proveedores_sin_repetido[value] = "proveedor_"+str(i)

for key, value in dict_proveedores_sin_repetido.items():
    print(f"key: {key} | value: {value}")
```

```python
df_main['proveedor'] = df_main['proveedor'].map(dict_proveedores_sin_repetido)
df_main
```
